```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
```

```python
# 나이 히스토그램
fig, ax = plt.subplots(figsize=(8,5))
titanic_nona = titanic.copy()
titanic_nona["Age"] = titanic_nona['Age'].fillna(titanic_nona['Age'].median())


ax.hist(
    titanic_nona['Age'],
    bins=20,
    color='skyblue',
    edgecolor='black'
)

ax.set_title('나이분포'),
ax.set_xlabel('Age')
ax.set_ylabel('Count')

plt.show()
```

```python
# ▪ 성별 생존자 막대
surv_class = titanic.groupby("Sex")["Survived"].sum()

fig, ax = plt.subplots(figsize=(7,5))
ax.bar(surv_class.index,surv_class.values,
       color=['#2563eb','#ea580c'], edgecolor='white')
ax.set_title('성별 생존자 막대')
ax.set_xticks([0, 1])
ax.set_xticklabels(["남성", "여성"])
plt.tight_layout(); plt.show()
```

```python
# ▪ 나이 vs 요금 산점도
colors = titanic_nona['Survived'].map({0: 'gray', 1: 'steelblue'})

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(titanic_nona['Age'], titanic_nona['Fare'],
           c=colors, alpha=0.4, s=30)
ax.set_xlabel('나이'); ax.set_ylabel('요금')
ax.set_title('나이 vs 요금 (색상=생존여부)')
plt.tight_layout(); plt.show()
```

```python
# ▪ 등급별 요금 박스플롯
fig, ax = plt.subplots(figsize=(7,5))

titanic.boxplot(
    column='Fare',
    by='Pclass',
    ax=ax
)

ax.set_title('등급별 요금 박스플롯')
ax.set_xlabel('객실등급(Pclass)')
ax.set_ylabel('요금(Fare)')

plt.suptitle('')  # 기본 제목 제거
plt.tight_layout()
plt.show()
```

```python
data_by_class = [titanic[titanic['Pclass']==c]['Fare'].dropna() for c in [1,2,3]]

fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot(data_by_class, tick_labels=['1등급','2등급','3등급'])
ax.set_ylabel('요금'); ax.set_title('등급별 요금 분포')
plt.tight_layout(); plt.show()
```

```python
# ▪ 탑승항 파이차트
port_counts = (titanic["Embarked"].value_counts().sort_index())

fig, ax = plt.subplots(figsize=(6,6))
ax.pie(port_counts.values, labels=port_counts.index,
       autopct='%1.1f%%', startangle=90,
       wedgeprops=dict(edgecolor='white')
       )
ax.set_title('탑승항별 승객 비율')
plt.tight_layout(); plt.show()
```

```python
# # plotly
# ▪ IMDB 평점 histogram
imdb = pd.read_csv('imdb_top_1000.csv')
imdb['Released_Year'] = pd.to_numeric(imdb['Released_Year'], errors='coerce')
```

```python
# ▪ 투표 수 vs 평점 scatter
# ▪ HTML 저장
import plotly.express as px

fig = px.scatter(
    imdb, x='No_of_Votes', y='IMDB_Rating',
    hover_name='Series_Title',            # 마우스 올리면 영화 제목
    hover_data=['Released_Year', 'Genre'],
    title='투표 수 vs IMDB 평점',
    template='simple_white', opacity=0.6
)
fig.show()
fig.write_html('imdb_scatter.html')

```

```python
# Genre = 'Crime, Drama' 형태 → 쪼개서 분석
imdb_genre = imdb.copy()
imdb_genre['Genre'] = imdb_genre['Genre'].str.split(',')
imdb_genre = imdb_genre.explode('Genre')
imdb_genre['Genre'] = imdb_genre['Genre'].str.strip()

# C1. 장르별 평균 IMDB 평점 수평 막대 (10편 이상만)
# C2. 연도별 평균 평점 추이 scatter
# C3. Gross 전처리 후 흥행 vs 평점 scatter
#     imdb['Gross_clean'] = imdb['Gross'].str.replace(',','', regex=False).astype(float)
```

```python
# C1. 장르별 평균 IMDB 평점 수평 막대 (10편 이상만)
genre_summary = (
    imdb_genre.groupby('Genre').agg(
        평균평점=('IMDB_Rating',"mean"),
        영화수=("Series_Title", "count")
    )
    .reset_index()
)

genre_summary = genre_summary[genre_summary['영화수'] >= 10]

genre_summary = genre_summary.sort_values('평균평점')

fig = px.bar(
    genre_summary,
    x='평균평점',
    y='Genre',
    orientation='h',
    title='장르별 평균 IMDB 평점 (10편 이상)',
    hover_data=['영화수'],
    template='simple_white'
)

fig.show()
fig.write_html('C1_genre_rating.html')
```

```python
# C2. 연도별 평균 평점 추이 scatter


year_rating = imdb.copy()

year_rating['Released_Year'] = pd.to_numeric(
    year_rating['Released_Year'],
    errors='coerce'
)

year_rating = year_rating.dropna(subset=['Released_Year'])

year_rating['Released_Year'] = year_rating['Released_Year'].astype(int)

year_summary = (
    year_rating.groupby('Released_Year')['IMDB_Rating']
    .mean()
    .reset_index()
)

fig = px.scatter(
    year_summary,
    x='Released_Year',
    y='IMDB_Rating',
    title='연도별 평균 IMDB 평점 추이',
    template='simple_white',
    opacity=0.7
)

fig.show()
fig.write_html('C2_year_rating.html')
```

```python
# C3. Gross 전처리 후 흥행 vs 평점 scatter

gross_df = imdb.copy()

gross_df['Gross_clean'] = (
    gross_df['Gross']
    .astype(str)
    .str.replace(',', '', regex=False)
)

gross_df['Gross_clean'] = pd.to_numeric(
    gross_df['Gross_clean'],
    errors='coerce'
)

gross_df = gross_df.dropna(subset=['Gross_clean', 'IMDB_Rating'])

fig = px.scatter(
    gross_df,
    x='Gross_clean',
    y='IMDB_Rating',
    hover_name='Series_Title',
    hover_data=['Released_Year', 'Genre'],
    title='흥행 수익 vs IMDB 평점',
    template='simple_white',
    opacity=0.6
)

fig.show()
fig.write_html('C3_gross_rating.html')
```

```python
import pandas as pd
import plotly.express as px 
```

```python
sa = pd.read_csv('seoul.csv',
encoding='cp949')
```

```python
sa.columns.tolist()
```

```python
sa['물건금액(만원)'] = (
    sa['물건금액(만원)']
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)

sa['계약일'] = sa['계약일'].astype(str)
```

```python
sa.head()
```

```python
# A1. 월별 평균 거래가격 꺾은선그래프 (HTML 저장)
monthly_price = (
    sa.groupby('계약일')['물건금액(만원)']
    .mean()
    .reset_index()
)

fig = px.line(
    monthly_price,
    x='계약일',
    y='물건금액(만원)',
    title='서울 아파트 월별 평균 거래가격',
    template='simple_white'
)

fig.update_xaxes(tickangle=45)
fig.show()
fig.write_html('A1_seoul_monthly_price.html')
```

```python
# A2. 월별 거래 건수 막대그래프 (HTML 저장)
monthly_count = (
    sa.groupby('계약일')
    .size()
    .reset_index(name='거래건수')
)

fig = px.bar(
    monthly_count,
    x='계약일',
    y='거래건수',
    title='월별 거래 건수',
    template='simple_white'
)

fig.update_xaxes(tickangle=45)

fig.show()
fig.write_html('A2_monthly_count.html')
```

```python
# A3. 구별 평균 거래가격 수평 막대그래프
gu_price = (
    sa.groupby('자치구명')['물건금액(만원)']
    .mean()
    .sort_values()
    .reset_index()
)

fig = px.bar(
    gu_price,
    x='물건금액(만원)',
    y='자치구명',
    orientation='h',
    title='구별 평균 거래가격',
    template='simple_white'
)

fig.show()
fig.write_html('A3_gu_price.html')
```
