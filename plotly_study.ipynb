```python
# 막대차트 그리기
# 결측치 제거해서 사본 만들기
# 결측 제거 + 복사본
df = imdb.dropna(subset=['Released_Year']).copy()

# 숫자형 변환
df['Released_Year'] = pd.to_numeric(
    df['Released_Year'],
    errors='coerce'
)

# 다시 결측 제거
df = df.dropna(subset=['Released_Year'])

# 정수형 변환
df['Released_Year'] = df['Released_Year'].astype(int)
```

```python
year_count = (
    df['Released_Year']
    .value_counts()
    .sort_index()
)

fig = px.bar(
    x=year_count.index,
    y=year_count.values,
    title='연도별 영화 수',
    labels={
        'x':'개봉연도',
        'y':'영화 수'
    },
    template='simple_white'
)

fig.show()
```
